```python
# Remover colunas que não irei utilizar
shark_df = shark.drop(['Case Number', 'Type', 'Location', 'Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order', 'Unnamed: 22', 'Unnamed: 23'], axis = 1)
```

```python
# transformando todos os valores em letra minuscula
shark_df['country']= shark_df['country'].str.lower().str.strip()
shark_df['injury']= shark_df['injury'].str.lower().str.strip()
shark_df['area']= shark_df['area'].str.lower().str.strip()
shark_df['activity']= shark_df['activity'].str.lower().str.strip()
shark_df['name']= shark_df['name'].str.lower().str.strip()
shark_df['fatal (y/n)']= shark_df['fatal (y/n)'].str.lower().str.strip()
```

```python
shark_df
```

```python
shark_df.loc[(shark_df['injury'] == 'fatal') & ~(shark_df['fatal (y/n)'] == 'y'),'fatal (y/n)']='y'
```

```python
shark_df = shark_df.dropna(subset='fatal (y/n)')
```

```python
shark_df = shark_df.dropna(subset='activity')
```

```python
shark_df = shark_df.dropna(subset='country')
```

```python
shark_mortes=shark_df.loc[shark_df['fatal (y/n)']=='y']
```

```python
shark_country=shark_mortes.value_counts('country').head(30).reset_index()
shark_country.rename(columns={0: 'mortes_por_pais'},inplace=True)
shark_country
```

```python
shark_activity=shark_mortes.value_counts('activity').head(30).reset_index()
shark_activity.rename(columns={0: 'mortes_por_atividade'},inplace=True)
shark_activity
```
